```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 300 dpi
plt.rcParams['figure.dpi'] = 300
```

```python
data_dir = './tex/analysis'
```

```python
# plot test accuracy against dataset size, where x-axis is the percentage of the dataset used
# and the y-axis is the test accuracy 
dl_data_df = pd.read_csv(os.path.join(data_dir, 'deeplearning_data_size.csv'))
hdc_data_df = pd.read_csv(os.path.join(data_dir, 'hdc_data_size.csv'))

plt.figure(figsize=(10, 5))
plt.plot(dl_data_df['Dataset Pct.'], dl_data_df['Accuracy'], label='Deep Learning')
plt.plot(hdc_data_df['Dataset Pct.'], hdc_data_df['Accuracy'], label='HDC')
plt.xlabel('Percentage of Dataset Used')
# use log scale for x-axis
plt.xscale('log')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Dataset Size')
plt.legend()
plt.savefig(os.path.join(data_dir, 'data_size.png'), bbox_inches='tight')
plt.show()
```

```python
# plot test accuracy against corruption rate, where x-axis is the percentage of the dataset used
# and the y-axis is the test accuracy 
dl_corruption_df = pd.read_csv(os.path.join(data_dir, 'deeplearning_corruption.csv'))
hdc_data_df = pd.read_csv(os.path.join(data_dir, 'hdc_corruption.csv'))

plt.figure(figsize=(10, 5))
plt.plot(dl_corruption_df['Corruption Rate'], dl_corruption_df['Accuracy'], label='Deep Learning')
plt.plot(hdc_data_df['Corruption Rate'], hdc_data_df['Accuracy'], label='HDC')
plt.xlabel('Corruption Rate')
# use log scale for x-axis
# plt.xscale('log')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Corruption Rate')
plt.legend()
plt.savefig(os.path.join(data_dir, 'corruption.png'), bbox_inches='tight')
plt.show()
```

```python
# speed analysis bar chart, load first column (Model) as index
speed_df = pd.read_csv(os.path.join(data_dir, 'speed_analysis.csv'), index_col=0)
speed_df.T.plot.bar(figsize=(10, 5))
plt.xlabel('Model')
plt.ylabel('Time (s)')
plt.title('Time to Train and Test Models')
plt.savefig(os.path.join(data_dir, 'speed.png'), bbox_inches='tight')
plt.show()
```

```python
# just look at Training-Time
# speed analysis bar chart, load first column (Model) as index
# different color for each model
colors = ['red', 'green', 'blue'] 
speed_df.T.loc['Training-Time'].plot.bar(figsize=(10, 5), color=colors)
plt.xlabel('Model')
# rotate x-axis labels 45 degrees
plt.xticks(rotation=45)
plt.ylabel('Time (s)')
plt.title('Time to Train Models')
plt.savefig(os.path.join(data_dir, 'speed_train.png'), bbox_inches='tight')
plt.show()
```

```python
# just look at Training-Time
# speed analysis bar chart, load first column (Model) as index
colors = ['red', 'green', 'blue'] 
speed_df.T.loc['Testing-Time'].plot.bar(figsize=(10, 5), color=colors)
plt.xlabel('Model')
# rotate x-axis labels 45 degrees
plt.xticks(rotation=45)
plt.ylabel('Time (s)')
plt.title('Time to Test Models')
plt.savefig(os.path.join(data_dir, 'speed_test.png'), bbox_inches='tight')
plt.show()
```
